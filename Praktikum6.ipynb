**Modul 6 : Stacked Vertkal Bar Chart**

**1. Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**2. Dataset**

```python
# Dataset
stores = ['Store A', 'Store B', 'Store C']
male_population = [150, 200, 180]
female_population = [120, 230, 170]

q1_male = [150, 180, 160]
q1_female = [140, 200, 180]
q2_male = [170, 190, 175]
q2_female = [130, 210, 160]
q3_male = [160, 185, 170]
q3_female = [135, 195, 165]
q4_male = [155, 175, 168]
q4_female = [125, 205, 158]
```

**3. Basic Stacked Vertical Bar Chart**

```python
x = np.arange(len(stores))
plt.bar(x, male_population, label='Male', color='blue')
plt.bar(x, female_population, bottom=male_population, label='Female', color='pink')

plt.xlabel('Stores')
plt.ylabel('Population')
plt.title('Population by Gender and Store')
plt.xticks(x, stores)
plt.legend()
plt.show()
```

**4. Kustomisasi Stacked Vertical Bar Chart**

```python
x = np.arange(len(stores))
plt.bar(x, male_population, label='Male', color='blue')
plt.bar(x, female_population, bottom=male_population, label='Female', color='pink')

plt.xlabel('Stores')
plt.ylabel('Population')
plt.title('Population by Gender and Store')
plt.xticks(x, stores)
plt.legend()

for i in range(len(x)):
    plt.text(x[i], male_population[i] / 2,
             str(male_population[i]), ha='center', color='white')
    plt.text(x[i], male_population[i] + (female_population[i] / 2),
             str(female_population[i]), ha='center', color='black')

plt.show()
```

**5. Multiple Stacked Vertical Bar Chart**

```python
x = np.arange(len(stores))
bar_width = 0.2
plt.bar(x - bar_width * 1.5, q1_male, label='Q1 Male', width=bar_width)
plt.bar(x - bar_width * 1.5, q1_female, bottom=q1_male, label='Q1 Female', width=bar_width)
plt.bar(x - bar_width / 2, q2_male, label='Q2 Male', width=bar_width)
plt.bar(x - bar_width / 2, q2_female, bottom=q2_male, label='Q2 Female', width=bar_width)
plt.bar(x + bar_width / 2, q3_male, label='Q3 Male', color='green', width=bar_width)
plt.bar(x + bar_width / 2, q3_female, bottom=q3_male, label='Q3 Female', color='yellow', width=bar_width)
plt.bar(x + bar_width * 1.5, q4_male, label='Q4 Male', color='purple', width=bar_width)
plt.bar(x + bar_width * 1.5, q4_female, bottom=q4_male, label='Q4 Female', color='orange', width=bar_width)

plt.xlabel('Stores')
plt.ylabel('Population')
plt.title('Population by Gender and Store (Quarterly)')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
